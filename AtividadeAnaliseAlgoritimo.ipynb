> Desenvolva um programa capaz de recuperar o índice da linha e o índice da coluna de um elemento em uma matriz triangular inferior quadrada, dada uma chave (valor) a ser buscada na matriz.

>  Observações:

–       A quantidade de linhas e colunas (ordem) deve ser definida pelo usuário do programa.

–       Preencha os valores da matriz por uma função que retorna valores aleatórios inteiros entre 1 e 9.

```python
def criar_matriz_triangular_inferior(ordem):
  matriz = []
  for i in range(ordem):
    linha = []
    for j in range(ordem):
      if i >= j:
        linha.append(random.randint(1, 9))
      else:
        linha.append(0)
    matriz.append(linha)
  return matriz

def buscar_elemento(matriz, valor):
  ordem = len(matriz)
  for i in range(ordem):
    for j in range(ordem):
      if i >= j and matriz[i][j] == valor:
        return (i+1, j+1)
  return None

ordem = int(input("Informe a ordem da matriz: "))
matriz = criar_matriz_triangular_inferior(ordem)
print(matriz)
valor = int(input("Informe o valor a ser buscado na matriz: "))
resultado = buscar_elemento(matriz, valor)
if resultado is not None:
  print("Índice da linha:", resultado[0])
  print("Índice da coluna:", resultado[1])
else:
  print("Valor não encontrado na matriz.")
```

**Você consegue identificar quais são o pior caso e melhor caso do algoritmo para encontrar o índice da linha e coluna (ordem) do valor pesquisado?**

*resposta:*

 O pior caso do algoritmo é quando o elemento procurado não existe na matriz, nesse caso o algoritmo precisa percorrer todos os elementos da matriz sem encontrar a chave, levando tempo de complexidade O(n^2), sendo n a ordem da matriz.

Já o melhor caso do algoritmo é quando o elemento procurado é encontrado na primeira iteração, ou seja, na primeira posição (0,0) da matriz, nesse caso o tempo de complexidade é O(1). Em geral, quanto mais cedo o elemento for encontrado, menor será o tempo de execução do algoritmo.

 **> Qual é a notação “big O” (ordem de complexidade) do algoritmo?**

*resposta:*

A notação "big O" (ordem de complexidade) desse algoritmo é O(n^2), onde n é a ordem da matriz triangular inferior. Isso se dá porque o algoritmo percorre todas as linhas e colunas da matriz para procurar o elemento, o que leva a uma complexidade quadrática.


**> Qual a curva de tempo de execução do algoritmo implementado para valores n igual a 3 até n igual a 100, onde n é o número da ordem da matriz triangular inferior quadrada.**

```python
import random
import time
import matplotlib.pyplot as plt

def gerar_matriz_triangular_inferior(ordem):
    matriz = []
    for i in range(ordem):
        linha = []
        for j in range(ordem):
            if i >= j:
                linha.append(random.randint(1, 9))
            else:
                linha.append(0)
        matriz.append(linha)
    return matriz

def buscar_elemento(matriz, valor):
    ordem = len(matriz)
    for i in range(ordem):
        for j in range(ordem):
            if matriz[i][j] == valor:
                return (i, j)
    return None

ordem_min = 3
ordem_max = 100
tempos = []
ordens = []
valor = 5

for ordem in range(ordem_min, ordem_max + 1):
    matriz = gerar_matriz_triangular_inferior(ordem)
    inicio = time.time()
    buscar_elemento(matriz, valor)
    fim = time.time()
    tempos.append(fim - inicio)
    ordens.append(ordem)

plt.plot(ordens, tempos)
plt.xlabel("Ordem da Matriz")
plt.ylabel("Tempo de Execução (segundos)")
plt.title("Curva de Tempo de Execução")
plt.show()
```
